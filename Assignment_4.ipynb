
###  Data Visualization - Pie Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
file = pd.read_csv("C:\\Users\\Laxmikant\\OneDrive\\Desktop\\airquality1.csv")
file
```

```python
file["Ozone"]= file["Ozone"].fillna(file["Ozone"].mean())
file["Solar.R"]= file["Solar.R"].fillna(file["Solar.R"].mean())
```

```python
file
```

```python
sizes = [file['Ozone'].mean(),file['Solar.R'].mean(),file['Wind'].mean(),file['Temp'].mean()]
colors = ['purple','greenyellow','blue','red']
labels = 'Ozone','Solar.R','Wind','Temp'
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
plt.title("Average Data")
```

```python
h = file.iloc[1:21,3]
h
```

### Bar graph 

```python
import numpy as np
y_posi=np.arange(len(h))
v=range(1,21)
```

```python
plt.bar(y_posi, height=h, width=0.8,color="purple")
plt.xticks(y_posi,v)
plt.title('Temperature of 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Temprature')
```

###  Data Visualization - BOXPLOT

```python
file.boxplot(column = ['Ozone','Wind'],)
plt.title('Box-Plot')
plt.suptitle('Temperature')
plt.ylabel('range')
```

### Histogram

```python
h = file.iloc[:,1]
```

```python
plt.hist(h, color='greenyellow')
plt.title('Histogram')
plt.ylabel('Range')
plt.xlabel('Days')
```

### Line graph


```python
plt.plot(file.iloc[0: 30, 6], file.iloc[0:30, 4], color="blue")
plt.title("Temp of May month")
plt.xlabel("Day")
plt.ylabel("Temp")
```

```python
plt.scatter(file.iloc[0:30, 6], file.iloc[0: 30, 3], color="red")
plt.title("Wind of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```

```python
with open("test.txt","r") as file:
    text=file.read()
print(text)
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords=set(STOPWORDS)
```

```python
wordcloud=WordCloud(width=3500,height=3500,background_color='white',stopwords=stopwords,min_font_size=20).generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
```
